```python
%%writefile Factorial.c
#include <stdio.h>

int fact_topdown(int n) {
    if (n == 0 || n == 1) return 1;
    return n * fact_topdown(n - 1);
}

int fact_bottomup_helper(int n, int curr, int res) {
    if (curr > n) return res;
    return fact_bottomup_helper(n, curr + 1, res * curr);
}

int fact_bottomup(int n) {
    return fact_bottomup_helper(n, 1, 1);
}

int fact_iterative(int n) {
    int res = 1;
    for (int i = 1; i <= n; i++) res *= i;
    return res;
}

int main() {
    int n;
    printf("Enter a number: ");
    scanf("%d", &n);

    printf("Top-Down Recursion: %d\n", fact_topdown(n));
    printf("Bottom-Up Recursion: %d\n", fact_bottomup(n));
    printf("Iterative Approach: %d\n", fact_iterative(n));

    return 0;
}
```
